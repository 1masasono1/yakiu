```python
df_train_pitch2 =  df_train_pitch.drop(df_train_pitch.columns[[0,3,42,43,44,45,46,47,48]], axis=1)
```

```python
df_train_pitch2
```

```python
print(df_train_pitch2['プレイ前ストライク数'].str.cat(df_train_pitch2['プレイ前ボール数'], sep='-'))
```

```python
df_train_pitch2['count'] = df_train_pitch2[['プレイ前ストライク数', 'プレイ前ボール数', 'プレイ前アウト数']].apply(lambda x: '{}-{}-{}'.format(x[0], x[1], x[2]), axis=1)
```

```python
df_train_pitch2['count']
```

```python
df_train_pitch2.info()
```

```python
df_train_pitch3 = df_train_pitch2.drop(df_train_pitch2.columns[[32,33,34]], axis=1)
```

```python
df_train_pitch3['pitch_team_score'] = 100
```

```python
i = range(len(df_train_pitch3))

for x in i:
    if df_train_pitch3.loc[x,'表裏'] == '表':
        df_train_pitch3.loc[x,'pitch_team_score'] = df_train_pitch3.loc[x,'プレイ前ホームチーム得点数']-df_train_pitch3.loc[x,'プレイ前アウェイチーム得点数']
    else:
                df_train_pitch3.loc[x,'pitch_team_score'] = df_train_pitch3.loc[x,'プレイ前アウェイチーム得点数']-df_train_pitch3.loc[x,'プレイ前ホームチーム得点数']
```

```python
pd.to_pickle(df_train_pitch3,'./df_train_pitch3.pkl')
```

```python
df_train_pitch3 = pd.read_pickle('./df_train_pitch3.pkl')
```

```python
df_date =df_train_pitch3['日付'].str.split('-', expand=True)
df_date
```

```python
df_date.rename(columns={0: 'year', 1: 'month',2:'day'}, inplace=True)
df_date = df_date.loc[:,['month']]
df_date = df_date.astype({'month':int})
df_date.head()
```

```python
df_train_pitch3 = pd.concat([df_train_pitch3,df_date],axis=1)
```

```python
df_train_pitch3.head(100)
```

```python
df_train_pitch3.dtypes
```

```python
df_train_pitch3['球場名'].most_common()
```

```python
pd.crosstab(df_train_pitch3['球場名'])
```

```python
pitch_group = df_train_pitch3.groupby(['試合ID','イニング','表裏','イニング内打席数'])
```

```python
lag_pitch = pd.DataFrame(pitch_group['球種'].shift(1))
```

```python
lag_pitch = lag_pitch.fillna(15)
lag_pitch = lag_pitch.rename(columns={'球種': 'lag_pitch'})
lag_pitch['lag_pitch'] = lag_pitch['lag_pitch']
lag_pitch = lag_pitch.astype('int64')
lag_pitch
```

```python
df_train_pitch4 = pd.concat([df_train_pitch3,lag_pitch],axis=1)
```

```python
df_train_pitch4.head(10)
```

```python
pd.to_pickle(df_train_pitch4,'./df_train_pitch4.pkl')
```
